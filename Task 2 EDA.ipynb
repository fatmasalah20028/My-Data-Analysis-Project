```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
df = pd.read_csv("C:/Users/AL-Majd/Desktop/Coveda Analysis Internship/cleaned_dataset.csv")
df.describe()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df['Retweets'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Retweets')

plt.subplot(1,2,2)
sns.histplot(df['Likes'], bins=15, kde=True, color='lightgreen')
plt.title('Distribution of Likes')

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.boxplot(y=df['Retweets'], color='skyblue')
plt.title('Boxplot of Retweets')

plt.subplot(1,2,2)
sns.boxplot(y=df['Likes'], color='lightgreen')
plt.title('Boxplot of Likes')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Retweets', y='Likes', data=df, color='purple')
plt.title('Retweets vs Likes')
plt.show()
```

```python
corr = df[['Retweets','Likes','Year','Month','Day','Hour']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation between numerical features')
plt.show()
```

```python
likes_per_year = df.groupby('Year')['Likes'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(data=likes_per_year, x='Year', y='Likes', marker='o', color='blue')
plt.title('Average Likes per Year')
plt.xlabel('Year')
plt.ylabel('Average Likes')
plt.grid(True)
plt.tight_layout()
plt.savefig('linechart_likes_year.png')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Retweets', y='Likes', data=df, hue='Platform', palette='Set2')
plt.title('Retweets vs Likes by Platform')
plt.xlabel('Retweets')
plt.ylabel('Likes')
plt.legend(title='Platform')
plt.tight_layout()
plt.savefig('scatter_retweets_likes.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Platform', y='Likes', data=df, palette='pastel')
plt.title('Distribution of Likes per Platform')
plt.xlabel('Platform')
plt.ylabel('Likes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('boxplot_likes_platform.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Retweets'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Retweets')
plt.xlabel('Retweets')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('histogram_retweets.png')
plt.show()
```
