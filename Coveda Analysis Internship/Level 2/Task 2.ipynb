```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv("C:/Users/AL-Majd/Desktop/Coveda Analysis Internship/cleaned_dataset.csv")


df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df = df.sort_values('Timestamp')
df.set_index('Timestamp', inplace=True)


ts = df['Likes']

plt.figure(figsize=(12,6))
plt.plot(ts)
plt.title('Time Series of Likes')
plt.xlabel('Date')
plt.ylabel('Likes')
plt.show()

```

```python
from statsmodels.tsa.seasonal import seasonal_decompose


decomposition = seasonal_decompose(ts, model='additive', period=30)
trend = decomposition.trend

plt.figure(figsize=(12,4))
plt.plot(trend)
plt.title('Trend Component of Likes')
plt.xlabel('Date')
plt.ylabel('Trend')
plt.show()
```

```python
seasonal = decomposition.seasonal

plt.figure(figsize=(12,4))
plt.plot(seasonal)
plt.title('Seasonal Component of Likes')
plt.xlabel('Date')
plt.ylabel('Seasonality')
plt.show()
```

```python
residual = decomposition.resid

plt.figure(figsize=(12,4))
plt.plot(residual)
plt.title('Residual Component of Likes')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.show()
```

```python
ma_7 = ts.rolling(window=7).mean()
ma_30 = ts.rolling(window=30).mean()

plt.figure(figsize=(12,6))
plt.plot(ts, label='Original Likes')
plt.plot(ma_7, label='7-Day Moving Average', color='red')
plt.plot(ma_30, label='30-Day Moving Average', color='green')
plt.title('Moving Average Smoothing')
plt.xlabel('Date')
plt.ylabel('Likes')
plt.legend()
plt.show()
```
